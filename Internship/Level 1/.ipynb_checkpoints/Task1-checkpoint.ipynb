```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Prithviraj\internship\Internship\Restaurant.csv")
```

```python
df.info()
```

```python
#I have used shape property to get number of rows and columns, and returened the value in two variables
rows,columns = df.shape


print("Number of rows:", rows)
print("Number of columns:", columns)
```

```python
# I have used this method which counts the number of missing values in each column.
missing_values = df.isnull().sum()

print("Missing values in each column:")
print(missing_values)
```

```python
print("Current data types:")
print(df.dtypes)
```

```python
print("Descriptive Statistics:")
print(df['Aggregate rating'].describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Aggregate rating', data=df)
plt.title("Box Plot of Aggregate Rating")
plt.ylabel("Aggregate Rating")
plt.show()
```

```python
# Trying to find the class imbalance in aggregate rating
class_counts = df['Aggregate rating'].value_counts()
print("Class Counts:")
print(class_counts)

#Following code will help to visualize the class imbalances 
plt.figure(figsize=(8, 6))
sns.countplot(x='Aggregate rating', data=df)
plt.title("Class Imbalance Check")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
df['Restaurant Name']=df['Restaurant Name'].replace('[�۱]+','',regex=True)
df['Address']=df['Address'].replace('[�۱]+','',regex=True)
df['Locality']=df['Locality'].replace('[�۱]+','',regex=True)
df['City']=df['City'].replace('[�۱]+','',regex=True)
df['Locality Verbose']=df['Locality Verbose'].replace('[�۱]+','',regex=True)
df['Cuisines']=df['Cuisines'].replace('[ï¿½]+','',regex=True)
```

```python
df['Restaurant Name Length'] = df['Restaurant Name'].apply(lambda x: len(x))
df['Address Length'] = df['Address'].apply(lambda x: len(x))

print(df[['Restaurant Name', 'Restaurant Name Length']])
print(df[['Address', 'Address Length']])
```
