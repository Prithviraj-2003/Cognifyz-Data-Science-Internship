```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Prithviraj\internship\Internship\Restaurant.csv")
```

```python
df.describe()
```

```python
#Distribution of Country Code is done below using seaborn and matplotlib libraries
plt.figure(figsize=(10, 6))
sns.countplot(x='Country Code', data=df)
plt.title("Distribution of Country Code")
plt.xlabel("Country Code")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(25, 6))
sns.countplot(x='City', data=df, order=df['City'].value_counts().index)
plt.title("Distribution of City")
plt.xlabel("City")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
#Here we are plotting for the top 20 cuisines

cuisines_series = df['Cuisines'].str.split(', ', expand=True).stack()
cuisine_counts = cuisines_series.value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_counts.index[:20], y=cuisine_counts.values[:20])
plt.title("Top 20 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```

```python
#Finding top cuisines and top cities with highest number of restaurant
cuisines_series = df['Cuisines'].str.split(', ', expand=True).stack()


cuisine_counts = cuisines_series.value_counts()

# Plotting top 10 cuisines , if we want the top 5 just change the value below in head(10) as head(5)
plt.figure(figsize=(12, 6))
plt.bar(cuisine_counts.head(10).index, cuisine_counts.head(10).values)
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()

top_cities = df['City'].value_counts().head(10)

#here we are plotting for top cities
plt.figure(figsize=(12, 6))
plt.bar(top_cities.index, top_cities.values)
plt.title("Top 10 Cities with the Highest Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=90)
plt.show()
```
