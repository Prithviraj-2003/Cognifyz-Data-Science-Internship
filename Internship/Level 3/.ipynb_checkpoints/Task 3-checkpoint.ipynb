```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df= pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Prithviraj\internship\Internship\Restaurant.csv")
```

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# below is the Histogram
axs[0].hist(df['Aggregate rating'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title('Histogram of Ratings')
axs[0].set_xlabel('Rating')
axs[0].set_ylabel('Frequency')

# below is the Bar Plot
rating_counts = df['Aggregate rating'].value_counts().sort_index()
axs[1].bar(rating_counts.index, rating_counts.values, color='salmon', edgecolor='black')
axs[1].set_title('Bar Plot of Ratings')
axs[1].set_xlabel('Rating')
axs[1].set_ylabel('Frequency')



plt.tight_layout()
plt.show()
```

```python
cuisine_series = df['Cuisines'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
cuisine_ratings_df = pd.DataFrame({'Cuisine': cuisine_series.values})
cuisine_ratings_df['Rating'] = df.loc[cuisine_series.index]['Aggregate rating'].values
average_rating_per_cuisine = cuisine_ratings_df.groupby('Cuisine')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=average_rating_per_cuisine.head(20).index, y=average_rating_per_cuisine.head(20).values, palette='viridis')
plt.title('Average Ratings for Top 20 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```
