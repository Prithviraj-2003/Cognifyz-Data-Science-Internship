Task: Feature Engineering

1] Extract additional features from the existing columns, such as the length of the restaurantname or address.

2] Create new features like "Has Table Booking"or "Has Online Delivery" by encodingcategorical variables.

```python
import pandas as pd

```

```python
df= pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Prithviraj\internship\Internship\Restaurant.csv")
```

```python
df['Restaurant Name Length'] = df['Restaurant Name'].apply(lambda x: len(x))
df['Address Length'] = df['Address'].apply(lambda x: len(x))

print(df[['Restaurant Name', 'Restaurant Name Length']])
print(df[['Address', 'Address Length']])
```

```python
df.head()
```

```python
mapping = {'Yes': 1, 'No': 0}
df['Has Table booking'] = df['Has Table booking'].map(mapping)
df['Has Online delivery'] = df['Has Online delivery'].map(mapping)

print(df)
```
